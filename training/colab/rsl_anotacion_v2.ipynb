```python
import re
from pypdf import PdfReader
from pathlib import Path

BASE_DIR = '/content'

def extraer_texto(pdf_path):
    try:
        reader = PdfReader(str(pdf_path))
        return ' '.join(
            (page.extract_text() or '').replace('\n', ' ')
            for page in reader.pages
        )
    except Exception as e:
        print(f'  ERROR {Path(pdf_path).name}: {e}')
        return ''

def encontrar_chunk(texto, output, tam=600):
    """
    Encuentra el fragmento de texto del PDF que mejor corresponde
    al output que anotaste manualmente.
    Si el output es SIN INFORMACION, devuelve un chunk central del articulo
    (para que el modelo aprenda que ese texto no contiene esa informacion).
    """
    palabras_pdf = texto.split()
    if not palabras_pdf:
        return output[:600]

    es_negativo = output.strip().upper().startswith('[SIN INFORMACION]')

    if es_negativo:
        # Para negativos: chunk del 30%-60% del articulo (metodologia/resultados)
        inicio = int(len(palabras_pdf) * 0.30)
        return ' '.join(palabras_pdf[inicio:inicio + tam])

    # Para positivos: buscar la ventana con mayor solapamiento de palabras con el output
    palabras_output = set(re.findall(r'\w{4,}', output.lower()))  # palabras de 4+ letras
    palabras_output -= {'este', 'esta', 'para', 'pero', 'como', 'more', 'than',
                        'with', 'that', 'from', 'have', 'were', 'which', 'been'}

    step    = tam // 2
    ventanas = [
        ' '.join(palabras_pdf[i:i + tam])
        for i in range(0, max(1, len(palabras_pdf) - tam), step)
    ]
    if not ventanas:
        return texto[:600]

    mejor = max(
        ventanas,
        key=lambda v: len(palabras_output & set(re.findall(r'\w{4,}', v.lower())))
    )
    return mejor

print('Funciones OK')
```

```python
import json

dataset  = []
sin_pdf  = []

for anotacion in ANOTACIONES:
    pdf_name = anotacion['pdf']
    dominio  = anotacion['dominio']
    pdf_path = Path(BASE_DIR) / dominio / pdf_name

    if not pdf_path.exists():
        print(f'[NO ENCONTRADO] {pdf_path}')
        sin_pdf.append(pdf_name)
        continue

    texto = extraer_texto(pdf_path)
    if len(texto.split()) < 100:
        print(f'[POCO TEXTO] {pdf_name}')
        continue

    for columna, output in anotacion['columnas'].items():
        chunk = encontrar_chunk(texto, output)
        dataset.append({
            'chunk':   chunk,
            'column':  columna,
            'output':  output,
            'dominio': dominio,
            'fuente':  pdf_name,
        })

pos = sum(1 for x in dataset if '[SIN INFORMACION]' not in x['output'].upper())
neg = len(dataset) - pos

print(f'\n=== DATASET GENERADO ===')
print(f'Total ejemplos  : {len(dataset)}')
print(f'Con informacion : {pos}')
print(f'Sin informacion : {neg}')
print(f'Ratio           : {pos/max(1,neg):.1f}:1')
if sin_pdf:
    print(f'PDFs no encontrados ({len(sin_pdf)}): {sin_pdf}')
```
